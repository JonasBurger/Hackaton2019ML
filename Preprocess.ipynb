```
import pandas as pd
url = 'https://raw.githubusercontent.com/ChristophRaab/MLH/master/dataset.csv'
df1 = pd.read_csv(url)
# Dataset is now stored in a Pandas Dataframe
```

```
from sklearn import preprocessing

def normalize(data):
    scaler = preprocessing.MinMaxScaler()
    scaled = scaler.fit_transform(data)
    return pd.DataFrame(scaled)    
```

# Preprocessing

```
from sklearn import preprocessing
from enum import Enum

class EncoderType(Enum):
    OneHot = 0,
    Label = 1,
    Binary = 2

    
categorics =[('job', EncoderType.OneHot), ('marital', EncoderType.OneHot), ('education', EncoderType.OneHot), ('default', EncoderType.Binary),
             ('housing', EncoderType.OneHot), ('loan', EncoderType.Binary), ('contact', EncoderType.OneHot), ('month', EncoderType.Label),
             ('poutcome', EncoderType.OneHot), ('deposit', EncoderType.Binary)]

encoders = {}

for column in categorics:
    if column[1] == EncoderType.OneHot:
        encoders[column[0]] = preprocessing.OneHotEncoder()
    elif column[1] == EncoderType.Label:
        encoders[column[0]] = preprocessing.LabelEncoder()
    elif column[1] == EncoderType.Binary:
        encoders[column[0]] = preprocessing.LabelBinarizer()
        
    encoders[column[0]].fit(df1[column[0]].values.reshape(-1, 1))
    
```

```
cat_data = {}

for column in categorics:
    if column[1] == EncoderType.Label:
        cat_data[column[0]] = normalize(encoders[column[0]].transform(df1[column[0]].values.reshape(-1, 1)).reshape(-1,1))
    else:
        cat_data[column[0]] = encoders[column[0]].transform(df1[column[0]].values.reshape(-1, 1))
```

```
norm_col = ['age', 'balance', 'day', 'campaign', 'pdays', 'previous']

norm_data = {}

for column in norm_col:
    norm_data[column] = normalize(df1[column].values.reshape(-1,1))
    
```

```
combined_data = {**cat_data, **norm_data}
combined_data
```
